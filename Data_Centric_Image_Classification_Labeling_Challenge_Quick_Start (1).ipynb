## Basics

Both the original benchmark and the present challenge are structured around two primary stages: Labeling and Evaluation.

````python
import torch
print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0))
````

outline-0.png


The original paper describes these phases as follows:

>"The main structure of the benchmark is divided into a Labeling and an Evaluation phase (see figure above), which is comparable to established Teacher-Student-Approaches. Using this denotation, the benchmarked `method` will, as a teacher, improve labels during the first phase (see the green box in the figure). These are then benchmarked in the second phase by analyzing their quality as training input to a student model (see the blue box in the figure). Be aware that we do not allow knowledge transfer from the second phase to the first phase.

>In detail, during the Labeling phase, we use samples from the distribution of the above-mentioned annotations to obtain different realistic label estimates as an `initialization` (see the orange box in the figure). The task of the benchmarked `method` (green) is to improve these estimates for better performance of another classification model (blue) in the second phase. In that phase (Evaluation), the obtained labels are used as input for training a fixed model, and its performance is measured on a testing subset of the original data."

In this challenge, participants can influence the `initialization` (depicted in orange) and the actual `method` for label refinement (illustrated in green). However, the evaluation model (represented in blue) remains immutable and beyond modification.

The primary objective of this challenge is to devise a labeling strategy encompassing both the `initialization` and `method` phases—depicted in orange and green, respectively—that generates labels for four distinct datasets. These labels are subsequently fed into a fixed but different evaluation models for training, the details of which are provided below. The ultimate performance evaluation hinges on the test-set performance of the trained models (depicted in blue) utilizing the labels furnished by participants.

Both the `initialization` and `method`  can use human annotations at a cost.  These annotations are procured from the `oracle`. The cost is encapsulated within the `budget`, which remains constant for each annotation and is delineated relative to the total dataset size, denoted as $N$. For instance, querying 10 annotations for 3 images via the `oracle` would entail a total `budget` expenditure of $10 \cdot 3/N$.

The evaluation metric employed is the distribution divergence between the predictions of the fixed model in the second phase and the averaged human perception. The  [Kullback-Leibler Divergence](https://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence)  (`KL`) to measure the difference between the distributions.

In summary, the overarching goal of this challenge is to devise an `initialization` and `method` strategy with a cumulative `budget` not exceeding 1. Furthermore, the mean `KL` across all four datasets should be minimized to achieve optimal performance.
